```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words("english"))

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|<.*?>|[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return " ".join(word for word in text.split() if word not in stop_words)


url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"
model = pd.read_csv(url, sep="\t", names=["Label", "Message"]).head(20)


model = model.dropna().drop_duplicates()


model["Message"] = model["Message"].apply(clean_text)
```

```python
from transformers import pipeline

classifier = pipeline("text-classification")

predicted_labels = []

for text in model["Message"]:
    result = classifier(text[:512])  # truncate long text to 512 tokens
    predicted_labels.append(result[0]['label'])

model["Predicted"] = predicted_labels
```
